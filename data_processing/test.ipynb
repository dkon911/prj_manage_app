```python
df.columns = df.columns.str.strip()

df = df[[
    'Deal Name', 'Project Type', 'Deal Amount',
    'Deal Received(MM/DD/YY)', 'Proposal Sent', 'Pending', 'Lost/Canceled', 'Won',
    'Division', 'Division 1 - %', 'Division 2 - %'
]]

df = df.rename(columns={
    'Deal Name': 'deal_name',
    'Project Type': 'project_type',
    'Deal Amount': 'deal_amount',
    'Deal Received(MM/DD/YY)': 'deal_received_date',
    'Proposal Sent': 'proposal_sent_date',
    'Lost/Canceled': 'lost_date',
    'Won': 'won_date',
    'Pending': 'pending_date',
    'Division': 'division',
    'Division 1 - %': 'division_1_pct',
    'Division 2 - %': 'division_2_pct'
})

df.dropna(subset=['deal_name'], inplace=True)

# Chun ha d liu: parse datetime
for date_col in ['deal_received_date', 'proposal_sent_date', 'won_date', 'lost_date', 'pending_date']:
    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

# Xc nh trng thi gn nht
def determine_status(row):
    if pd.notnull(row['won_date']):
        return 'Won'
    if pd.notnull(row['lost_date']):
        return 'Lost'
    if pd.notnull(row['pending_date']):
        return 'Pending'
    if pd.notnull(row['proposal_sent_date']):
        return 'Proposal Sent'
    if pd.notnull(row['deal_received_date']):
        return 'Received'
    return 'Unknown'

df['status'] = df.apply(determine_status, axis=1)

# i kiu tin t nu c $
df['deal_amount'] = df['deal_amount'].replace('[\$,]', '', regex=True).astype(float)

df['deal_id'] = df.index + 1  # hoc to bng uuid
```
